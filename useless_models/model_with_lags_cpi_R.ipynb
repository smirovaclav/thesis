```r
library(dplyr)
library(readr)
library(restriktor)
library(car)
```

```r
df1 <- read_csv("data_processing/final_dataset/intermediate_data_per_employee_cze.csv", show_col_types = FALSE)
df2 <- read_csv("data_processing/final_dataset/intermediate_data_per_hour_cze.csv", show_col_types = FALSE)
```

```r
lag_model <- df1 %>%
  mutate(

    d_2020Q2 = ifelse(date == "2020Q2", 1, 0),
    d_2021Q1 = ifelse(date == "2021Q1", 1, 0),
    
    gw_lag1 = lag(gw, 1),
    gw_lag2 = lag(gw, 2),
    gw_lag3 = lag(gw, 3),
    gw_lag4 = lag(gw, 4),

    cf1_lag1 = lag(cf1, 1),
    cf1_lag2 = lag(cf1, 2),
    cf1_lag3 = lag(cf1, 3),
    cf1_lag4 = lag(cf1, 4),

    diff_cpi_lag1 = lag(diff_cpi, 1),
    diff_cpi_lag2 = lag(diff_cpi, 2),
    diff_cpi_lag3 = lag(diff_cpi, 3),
    diff_cpi_lag4 = lag(diff_cpi, 4),

    vu_lag1 = lag(vu, 1),
    vu_lag2 = lag(vu, 2),
    vu_lag3 = lag(vu, 3),
    vu_lag4 = lag(vu, 4),

    gpty_lag1 = lag(gpty, 1)
  ) %>%

  select(
    date, 
    gw,          
    gw_lag1, gw_lag2, gw_lag3, gw_lag4,
    cf1_lag1, cf1_lag2, cf1_lag3, cf1_lag4,
    diff_cpi_lag1, diff_cpi_lag2, diff_cpi_lag3, diff_cpi_lag4,
    vu_lag1, vu_lag2, vu_lag3, vu_lag4,
    gpty_lag1,
    d_2020Q2, d_2021Q1
  ) %>%

  na.omit()

wage_model1 <- lm(gw ~ gw_lag1 + gw_lag2 + gw_lag3 + gw_lag4 + 
                      cf1_lag1 + cf1_lag2 + cf1_lag3 + cf1_lag4 +
                      diff_cpi_lag1 + diff_cpi_lag2 + diff_cpi_lag3 + diff_cpi_lag4 +
                      vu_lag1 + vu_lag2 + vu_lag3 + vu_lag4 +
                      gpty_lag1 +
                      d_2020Q2 + d_2021Q1,
                      data = lag_model)

summary(wage_model1)
```

```r
# p-stat (sum)
linearHypothesis(wage_model1, "gw_lag1 + gw_lag2 + gw_lag3 + gw_lag4 = 0")
# p-stat (joint)
linearHypothesis(wage_model1, c("gw_lag1=0", "gw_lag2=0", "gw_lag3=0", "gw_lag4=0"))
```

```r
# p-stat (sum)
linearHypothesis(wage_model1, "cf1_lag1 + cf1_lag2 + cf1_lag3 + cf1_lag4 = 0")
# p-stat (joint)
linearHypothesis(wage_model1, c("cf1_lag1=0", "cf1_lag2=0", "cf1_lag3=0", "cf1_lag4=0"))
```

```r
# p-stat (sum)
linearHypothesis(wage_model1, "diff_cpi_lag1 + diff_cpi_lag2 + diff_cpi_lag3 + diff_cpi_lag4 = 0")
# p-stat (joint)
linearHypothesis(wage_model1, c("diff_cpi_lag1=0", "diff_cpi_lag2=0", "diff_cpi_lag3=0", "diff_cpi_lag4=0"))
```

```r
# p-stat (sum)
linearHypothesis(wage_model1, "vu_lag1 + vu_lag2 + vu_lag3 + vu_lag4 = 0")
# p-stat (joint)
linearHypothesis(wage_model1, c("vu_lag1=0", "vu_lag2=0", "vu_lag3=0", "vu_lag4=0"))
```

```r
# vertical Phillips curve in LR
my_constraints <- 'gw_lag1 + gw_lag2 + gw_lag3 + gw_lag4 + cf1_lag1 + cf1_lag2 + cf1_lag3 + cf1_lag4 = 1'
wage_model_restricted1 <- restriktor(wage_model1, 
                                    constraints = my_constraints)

summary(wage_model_restricted1)
```

```r
beta <- coef(wage_model_restricted1)
sum_gw <- sum(beta[grep("gw_lag", names(beta))])
sum_gw
sum_cf1 <- sum(beta[grep("cf1_lag", names(beta))])
sum_cf1 
sum_vu <- sum(beta[grep("vu_lag", names(beta))])
sum_vu
sum_catchup <- sum(beta[grep("diff_cpi_lag", names(beta))])
sum_catchup
sum_gpty <- sum(beta[grep("gpty_lag1", names(beta))])
sum_gpty
```
