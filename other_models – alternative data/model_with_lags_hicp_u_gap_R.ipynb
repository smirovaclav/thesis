```r
library(dplyr)
library(readr)
library(restriktor)
library(car)
```

```r
setwd("/Users/smirovaclav/Documents/blanchard_papery/czechia_data/master_thesis/")
```

```r
df1 <- read_csv("data_processing/final_dataset/intermediate_data_per_employee_cze.csv", show_col_types = FALSE)
df2 <- read_csv("data_processing/final_dataset/intermediate_data_per_hour_cze.csv", show_col_types = FALSE)
```

```r
lag_model <- df1 %>%
  mutate(
      
    d_2020Q2 = ifelse(date == "2020Q2", 1, 0),
    d_2021Q1 = ifelse(date == "2021Q1", 1, 0),
    
    gw_lag1 = lag(gw, 1),
    gw_lag2 = lag(gw, 2),
    gw_lag3 = lag(gw, 3),
    gw_lag4 = lag(gw, 4),

    cf1_lag1 = lag(cf1, 1),
    cf1_lag2 = lag(cf1, 2),
    cf1_lag3 = lag(cf1, 3),
    cf1_lag4 = lag(cf1, 4),

    diff_hicp_lag1 = lag(diff_hicp, 1),
    diff_hicp_lag2 = lag(diff_hicp, 2),
    diff_hicp_lag3 = lag(diff_hicp, 3),
    diff_hicp_lag4 = lag(diff_hicp, 4),

    u_gap_lag1 = lag(u_gap, 1),
    u_gap_lag2 = lag(u_gap, 2),
    u_gap_lag3 = lag(u_gap, 3),
    u_gap_lag4 = lag(u_gap, 4),

    gpty_lag1 = lag(gpty, 1)
  ) %>%

  select(
    date, 
    gw,          
    gw_lag1, gw_lag2, gw_lag3, gw_lag4,
    cf1_lag1, cf1_lag2, cf1_lag3, cf1_lag4,
    diff_hicp_lag1, diff_hicp_lag2, diff_hicp_lag3, diff_hicp_lag4,
    u_gap_lag1, u_gap_lag2, u_gap_lag3, u_gap_lag4,
    gpty_lag1,
    d_2020Q2, d_2021Q1
  ) %>%

  na.omit()

wage_model1 <- lm(gw ~ gw_lag1 + gw_lag2 + gw_lag3 + gw_lag4 + 
                      cf1_lag1 + cf1_lag2 + cf1_lag3 + cf1_lag4 +
                      diff_hicp_lag1 + diff_hicp_lag2 + diff_hicp_lag3 + diff_hicp_lag4 +
                      u_gap_lag1 + u_gap_lag2 + u_gap_lag3 + u_gap_lag4 +
                      gpty_lag1 +
                      d_2020Q2 + d_2021Q1,
                      data = lag_model)

summary(wage_model1)
```

```r
# p-stat (sum)
linearHypothesis(wage_model1, "gw_lag1 + gw_lag2 + gw_lag3 + gw_lag4 = 0")
# p-stat (joint)
linearHypothesis(wage_model1, c("gw_lag1=0", "gw_lag2=0", "gw_lag3=0", "gw_lag4=0"))
```

```r
# p-stat (sum)
linearHypothesis(wage_model1, "cf1_lag1 + cf1_lag2 + cf1_lag3 + cf1_lag4 = 0")
# p-stat (joint)
linearHypothesis(wage_model1, c("cf1_lag1=0", "cf1_lag2=0", "cf1_lag3=0", "cf1_lag4=0"))
```

```r
# p-stat (sum)
linearHypothesis(wage_model1, "diff_hicp_lag1 + diff_hicp_lag2 + diff_hicp_lag3 + diff_hicp_lag4= 0")
# p-stat (joint)
linearHypothesis(wage_model1, c("diff_hicp_lag1=0", "diff_hicp_lag2=0", "diff_hicp_lag3=0",  "diff_hicp_lag4=0"))
```

```r
# p-stat (sum)
linearHypothesis(wage_model1, "u_gap_lag1 + u_gap_lag2 + u_gap_lag3 + u_gap_lag4 = 0")
# p-stat (joint)
linearHypothesis(wage_model1, c("u_gap_lag1=0", "u_gap_lag2=0", "u_gap_lag3=0", "u_gap_lag4=0"))
```

```r
# vertical Phillips curve in LR
my_constraints <- 'gw_lag1 + gw_lag2 + gw_lag3 + gw_lag4 + cf1_lag1 + cf1_lag2 + cf1_lag3 + cf1_lag4 = 1'
wage_model_restricted1 <- restriktor(wage_model1, 
                                    constraints = my_constraints)

summary(wage_model_restricted1)
```

```r
beta <- coef(wage_model_restricted1)
sum_gw <- sum(beta[grep("gw_lag", names(beta))])
sum_gw
sum_cf1 <- sum(beta[grep("cf1_lag", names(beta))])
sum_cf1 
sum_u_gap <- sum(beta[grep("u_gap_lag", names(beta))])
sum_u_gap
sum_catchup <- sum(beta[grep("diff_hicp_lag", names(beta))])
sum_catchup
sum_gpty <- sum(beta[grep("gpty_lag1", names(beta))])
sum_gpty
```
